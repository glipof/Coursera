```python
'''
Function: Find most frequent k-mers in a string
Input   : String (seq), Integer (k = length)
Output  : List of Strings (kmers) 
'''
import sys
from collections import defaultdict

# Data input
seq = sys.argv[0]
k = sys.argv[1]

# Finding k-mers
kmers = defaultdict(int)

def FindKmer(seq, k):
    seq = seq.upper()
    assert set(seq).issubset(set(['A', 'C', 'T', 'G', 'N'])), 'Invalid bases'
    for i in range(0, len(seq)-(k-1)):
        kmers[seq[i:i+k]] += 1
    return [key for key, val in kmers.items()
            if val == max(kmers.values())]
```

```python
'''
Function: Convert DNA seq to its reverse complement counterpart
Input   : String (seq)
Output  : Strings (rc) 
''' 
def ReverseComplement(seq):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A', 'N': 'N'}
    rc = ''
    for base in seq:
        rc = complement[base] + rc
    return rc
```

```python
'''
Function : Naive pattern matching of all occurrences in a string
Input: Strings (pattern, text)
Output: List of Integers (occurrences) 
'''
def Naive(pattern, text):
    occurrences = []
    for i in range(len(text) - len(pattern) + 1):  # loop over alignments
        match = True
        for j in range(len(pattern)):           # loop over characters
            if text[i+j] != pattern[j]:            # compare characters
                match = False
                break
        if match:
            occurrences.append(i)         # all chars matched; record
    return occurrences

'''
Function : Naive pattern matching w/ RC of all occurrences
Input: Strings (p = pattern, t = text)
Output: List of Integers (occurrences) 
'''
def NaiveWithRC(pattern, text):
    occurrences = Naive(pattern, text)
    revP = ReverseComplement(pattern)
    if pattern == revP:
        return occurrences
    else:
        more_occurrences = Naive(revP, text)
        return occurrences + more_occurrences
```

```python
'''
Function : Find Kmer Clumps
Input: Strings (genome), Integers (k = kmers, L = length, t = times)
Output: List of Strings (kmer_clumps) 
'''
from collections import defaultdict, deque, Counter
from itertools import islice

def GetClumps(genome, k, L, t):
    length = len(genome) - k + 1
    indexes = defaultdict(list)
    result = set()

    for i in range(length):
        kterm = genome[i:i + k]

        # remove positions with distance > L
        while indexes[kterm] and i + k - indexes[kterm][0] > L:
            indexes[kterm].pop(0)

        # add current kterm position
        indexes[kterm].append(i)

        if len(indexes[kterm]) == t:
            result.add(kterm)

    return result
```

```python
# Out of curiousity, I asked ChatGPT how to optimize GetClumps method.
# It suggested the following:
# - Use defaultdict(deque) for faster insertion and removal positions
# - Change indexes[kterm].pop(0) to popleft() since using deques
# - Use Counter which can be more efficient than checkly indexes[kterm] length

from collections import defaultdict, deque, Counter

def GetClumpsDeque(genome, k, L, t):
    length = len(genome) - k + 1
    indexes = defaultdict(deque)
    result = set()
    counter = Counter()

    for i in range(length):
        kterm = genome[i:i + k]

        # remove positions with distance > L
        while indexes[kterm] and i + k - indexes[kterm][0] > L:
            indexes[kterm].popleft()
            counter[kterm] -= 1

        # add current kterm position
        indexes[kterm].append(i)
        counter[kterm] += 1

        if counter[kterm] == t:
            result.add(kterm)

    return result
```

```python
# Compare execute time of GetClumps vs GetClumpsDeque using Ecoli

file = open("./E_coli.txt")
ecoli_dna = file.read()

import time

# Time execution for GetClumps
start_time = time.time()

ecoli_clumps = GetClumps(ecoli_dna, 9, 500, 3)
print(ecoli_clumps)

end_time = time.time()
elapsed_time = end_time - start_time

# Print the elapsed time for GetClumps 
print("GetClumps Elapsed time: {:.2f} seconds".format(elapsed_time))

# Time execution for GetClumpsDeque
start_time = time.time()

ecoli_clumps = GetClumpsDeque(ecoli_dna, 9, 500, 3)
print(ecoli_clumps)

end_time = time.time()
elapsed_time = end_time - start_time

# Print the elapsed time for GetClumpsDeque
print("GetClumpsDeque Elapsed time: {:.2f} seconds".format(elapsed_time))
```
