```python
import random
import numpy as np

def count_occurrences(motifs):
    counts = {'A': [], 'C': [], 'G': [], 'T': []}
    k = len(motifs[0])

    for i in range(k):
        column = [motif[i] for motif in motifs]
        counts['A'].append(column.count('A'))
        counts['C'].append(column.count('C'))
        counts['G'].append(column.count('G'))
        counts['T'].append(column.count('T'))

    return counts

def ProfileMatrix(motifs):
    counts = count_occurrences(motifs)
    k = len(motifs[0])
    profile = {'A': [], 'C': [], 'G': [], 'T': []}

    for nucleotide in ['A', 'C', 'G', 'T']:
        for i in range(k):
            profile[nucleotide].append((counts[nucleotide][i] + 1) / (len(motifs) + 4))  # Adding pseudocounts

    return profile

def ProfileMostProbableKmer(text, k, profile):
    if type(profile) == str:
        profile = profile.split(" ")

    profile        = np.array(list(profile.values()))
    prob_high      = 0
    output_pattern = text[0: k]
    
    for start in range(len(text) - k + 1):
        end      = start + k
        pattern  = text[start: end]
        prob_pattern = 1
        for i in range(k):
            if   pattern[i] == 'A': prob_pattern = prob_pattern * profile[0, i]
            elif pattern[i] == 'C': prob_pattern = prob_pattern * profile[1, i]
            elif pattern[i] == 'G': prob_pattern = prob_pattern * profile[2, i]
            elif pattern[i] == 'T': prob_pattern = prob_pattern * profile[3, i]
                
        if prob_pattern > prob_high:
            prob_high      = prob_pattern
            output_pattern = pattern
            
    return output_pattern


# Stepik Week 4 - Randomized Motif Search Algorithm with Pseudocounts

# Input:  String dnas, integer k (kmer length) and integer t (sequences)
# Output: matrix profile

def RandomMotifSearch(dnas, k, t):
    if type(dnas) == str:
        dnas = dnas.split(" ")

    best_motifs = []
    motifs = []
    best_score = float('inf')

    # Randomly select kmer motifs from dnas
    for dna in dnas:
        start = random.randrange(len(dna) - k + 1)
        end   = start + k
        motif = dna[start: end]
        motifs.append(motif)
    best_motifs = motifs

    # print('Random selected kmers =', best_motifs)

    while True:
        profile = ProfileMatrix(motifs)
        motifs  = []
        
        for dna in dnas:
            motif = ProfileMostProbableKmer(dna, k, profile)
            motifs.append(motif)
  
        best_score    = Score(best_motifs)
        current_score = Score(motifs)
        
        if current_score < best_score:
            best_motifs = motifs
        else:
            return best_motifs

# Sample Test (Output = 'TCTCGGGG CCAAGGTG TACAGGCG TTCAGGTG TCCACGTG')
k = 8
t = 5
dnas='CGCCCCTCTCGGGGGTGTTCAGTAAACGGCCA GGGCGAGGTATGTGTAAGTGCCAAGGTGCCAG TAGTACCGAGACCGAAAGAAGTATACAGGCGT TAGATCAAGTTTCAGGTGCACGTCGGTGAACC AATCCACCAGCTCCACGTGCAATGTTGGCCTA'

print('Sample Randomized Motif =', ' '.join(str(e) for e in RandomMotifSearch(dnas, k, t)))

# TODO: Output does not match expected results
```

```python
# Stepik Week 4 - Gibbs Sampling Algorithm

# Input:  String dnas, integer k (kmer length), integer t (sequences), integer N (range)
# Output: matrix profile

def Random(motifs, profile):
    if type(profile) == str:
        profile = profile.split(" ")

    profile        = np.array(list(profile.values()))
    prob_all = []
    
    for motif in motifs:
        prob_motif = 1
        for i in range(len(motif)):
            if   motif[i] == 'A': prob_motif = prob_motif * profile[0, i]
            elif motif[i] == 'C': prob_motif = prob_motif * profile[1, i]
            elif motif[i] == 'G': prob_motif = prob_motif * profile[2, i]
            elif motif[i] == 'T': prob_motif = prob_motif * profile[3, i]
        prob_all.append(prob_motif)
    
    sum_prob = sum(prob_all)
    for i in range(len(motifs)):
        prob_all[i] = prob_all[i]/sum_prob
    
    choose_index = random.choices(range(len(motifs)), prob_all)[0]
    
    return(choose_index)


def GibbsSampling(dnas, k, t, N):
    if type(dnas) == str:
        dnas = dnas.split(" ")

    best_motifs = []
    motifs = []
    best_score = float('inf')

    # Randomly select kmer motifs from dnas
    for dna in dnas:
        start = random.randrange(len(dna) - k + 1)
        end   = start + k
        motif = dna[start: end]
        motifs.append(motif)

    best_motifs = motifs
    profile = ProfileMatrix(motifs)

    for loop in range(N):
        i = Random(motifs, profile)
        tmp_motifs = motifs[:1] + motifs[(i+1):]
        profile = ProfileMatrix(motifs)
        motifs[i] = ProfileMostProbableKmer(dnas[i], k, profile)
  
        best_score    = Score(best_motifs)
        current_score = Score(motifs)
        
        if current_score < best_score:
            best_motifs = motifs
            best_score = current_score
    
    return best_motifs

# Sample Test (Output = 'TCTCGGGG CCAAGGTG TACAGGCG TTCAGGTG TCCACGTG')

k = 8
t = 5
N = 100
dnas = 'CGCCCCTCTCGGGGGTGTTCAGTAACCGGCCA GGGCGAGGTATGTGTAAGTGCCAAGGTGCCAG TAGTACCGAGACCGAAAGAAGTATACAGGCGT TAGATCAAGTTTCAGGTGCACGTCGGTGAACC AATCCACCAGCTCCACGTGCAATGTTGGCCTA'

print('Sample w/ range of 100 =', GibbsSampling(dnas, k, t, N))
```

```python
# Stepik Week 4 - Gibbs Sampling Algorithm

# Solution adapted from Tanshree Kar in Stepik
# Input:  String dnas, integer k (kmer length), integer t (sequences), integer N (range)
# Output: matrix profile

def GibbsSampler(Dna, k, t, N):
    if type(Dna) == str:
        Dna = Dna.split(" ")

    best_motifs = []
    best_score = float('inf')

    motifs = []

    for i in range(t):
        index = random.randint(0, len(Dna[i]) - k)
        motifs.append(Dna[i][index:index+k])

    for step in range(N):
        i = random.randint(0, t-1)
        motifs_except_i = motifs[:i] + motifs[i+1:]
        profile = ProfileMatrix(motifs_except_i)
        motifs[i] = ProfileMostProbableKmer(Dna[i], k, profile)
        score = Score(motifs)

        if score < best_score:
            best_score = score
            best_motifs = motifs.copy()
    
    return best_motifs

# Sample Test (Output = 'TCTCGGGG CCAAGGTG TACAGGCG TTCAGGTG TCCACGTG')

k = 8
t = 5
N = 100
dnas = 'CGCCCCTCTCGGGGGTGTTCAGTAACCGGCCA GGGCGAGGTATGTGTAAGTGCCAAGGTGCCAG TAGTACCGAGACCGAAAGAAGTATACAGGCGT TAGATCAAGTTTCAGGTGCACGTCGGTGAACC AATCCACCAGCTCCACGTGCAATGTTGGCCTA'

print('Sample w/ range of 100 =', GibbsSampler(dnas, k, t, N))
```
