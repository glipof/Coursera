```python
# Stepik Exercise - Approximinate Pattern Count/ Compute Count

def ApproximatePatternCount(pattern, genome, d):
    count = 0
    sub_string = [None]*(len(genome)-len(pattern)+1)
    for i in range(len(genome)-len(pattern)+1):
        sub_string[i] = genome[i:i+len(pattern)]
        if HammingDistance(pattern, sub_string[i]) <= d:
            count += 1
    return count 

# Sample genome output should be = 4
pattern = 'GAGG'
genome = 'TTTAGAGCCTTCAGAGG'
mismatches = 2
print("Sample Approximate Pattern Count:", ApproximatePatternCount(pattern, genome, mismatches))

# Randomized dataset output
pattern = 'GGGCGAC'
genome = 'ACTGCTCCGGCGGGGCGACTTCGTACCGGTCGTGTGTAATATGTTCACCCCAAATCGGCGTGGAAATCCTTGCCTAGCTGTCATGAGTTCTTTATGGCATGTATATCCCGACTCGTAACGAGTGGTAGAAGCCTTGTTTCATAACGGGCAGATCTTTGCTGCAACCTTTACTAAATGATGAGAATGGATAGGCAAACCTATCTATAATAACCGGAGTATCAGATAGGTCGCCATCTGGAAAAACATACGGATCACATTTAAATGGGGTATACTTGATCCAATGACAAAACCTCATTCTACGACTAAGGTTGCTTCCCCTGACCTATCAGTAAACACGCTTC'
mismatches = 3
print("Randomized Approximate Pattern Count:", ApproximatePatternCount(pattern, genome, mismatches))

# Week 2 Quiz - Question 4
pattern = 'AA'
genome = 'TACGCATTACAAAGCACA'
mismatches = 1
print("Quiz Approximate Pattern Count:", ApproximatePatternCount(pattern, genome, mismatches))
```

```python
#Functions Number to Pattern & Pattern to Number

def Number2Symbol(index):
    nucleotide = {"A":0, "C":1, "G":2, "T":3}
    for k, v in nucleotide.items():
        if v == index:
            return k
       
def Number2Pattern(index, k):
    if k == 1:
        return Number2Symbol(index)
    prefix_index = index // 4
    r = index % 4 
    symbol = Number2Symbol(r)
    prefix_pattern = Number2Pattern(prefix_index, k-1)
    return "".join([prefix_pattern, symbol]) 

# Sample Input (Output = AGTC)
index = 45
k = 4
print(Number2Pattern(index, k))

nucleotide = {"A":0, "C":1, "G":2, "T":3}
def Pattern2Number(Pattern):
    if len(Pattern) == 0:
        return 0    
    symbol = Pattern[-1]
    prefix = Pattern[0:-1]
    return (4 * Pattern2Number(prefix) + nucleotide.get(symbol))

# Sample input (Output = 11):
print(Pattern2Number('AGT'))
```

```python
def ReverseComplement(s):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A', 'N': 'N'}
    t = ''
    for base in s:
        t = complement[base] + t
    return t

# Test Case RC (Answer = ACCGGGTTTT)
print(ReverseComplement('AAAACCCGGT'))
```

```python
# Stepik Exercise - Frequent Words With Mismatches

def FreqWordsMismatch(genome, k, d):
    freq_pattern = []   
    close = [0]*(4**k)
    freq_array = [0]*(4**k)
    
    # count all possible k-mers from Text in 4^k-sized list
    
    for i in range(len(genome)-k+1):
        neighborhood = neighbors(genome[i:i+k], d)
        for pattern in neighborhood:
            index = Pattern2Number(pattern)
            close[index] = 1
            
    # for each k-mer, check how many times it shows up in Text
    # with maximum mismatch, d
    
    for i in range(4**k):
        if close[i] == 1:
            pattern = Number2Pattern(i, k)
            freq_array[i] = ApproximatePatternCount(pattern, genome, d)
    max_count = max(freq_array)
    
    for i in range(4**k):
        if freq_array[i] == max_count:
            pattern = Number2Pattern(i, k)
            freq_pattern.append(pattern)
    return freq_pattern

def neighbors(pattern, d):
    if d == 0:
        return pattern
    if len(pattern) == 1:
        return {'A', 'C', 'G', 'T'}
        
    neighborhood = []
    nucleotide = ['A', 'C', 'G', 'T']
    suffix_neighbors = neighbors(pattern[1:], d)

    for text in suffix_neighbors:
        if HammingDistance(pattern[1:], text) < d:
            for e in nucleotide:
                neighborhood.append(e+text)
        else:
            neighborhood.append(pattern[0]+text)
    return neighborhood

# - Sample Input (Output: ATGC ATGT GATG)
print('Sample =', ' '.join(str(e) for e in FreqWordsMismatch('ACGTTGCATGTCGCATGATGCATGAGAGCT', 4, 1)))

# - Randomized Dataset
print('Randomized =', ' '.join(str(e) for e in FreqWordsMismatch('CCGCGCCCGCCCTATAAGCGCCCGCCCCGAAGCCGCGCCCTATCCGGGACGGACGGACGGACGGACCGCCCTATCGCCCTATAAGGGACAAGGGACGGACCGCCCCGCGCCCCGCTATAAGAAGCCGCCGCCGAAGCCGCTATGGACCTATCCGCGCCCTATCTATAAGCTATCTATGGACCGCCCGCCCCGCGCCGGACCCGAAGCGCCAAGCGCCCCGGGACAAGCCGCTATAAGCCGCCGCCGCCGAAGGGACAAGAAGCGCCCGCCAAGCGCCAAGAAGCGCCCGCCAAGCGCCCGCCCCGAAGCGCCCCGCGCCAAGGGACCCGCTATCCGAAGCGCCCTATCTATCTAT', 5, 3)))

# Code Challenge Test (ACG, 1 = CCG TCG GCG AAG ATG AGG ACA ACC ACT ACG)

print("d-Neighbors =", ' '.join(neighbors('ACG', 1)))
```

```python
# Stepik Exercise - Frequent Words With Mismatches  and Reverse Complements

from functools import reduce

def FreqWordsMismatchRC(genome, k, d):
    freq_pattern = []
    neighborhood = []
    
    # all k-mers (both fwd & rev) in Text with maximum d mismatches
    for i in range(len(genome)-k+1):
        reverse = ReverseComplement(genome[i:i+k])
        neighborhood.append(neighbors(genome[i:i+k], d))
        neighborhood.append(neighbors(reverse, d))
    neighborhood_array = reduce(lambda x, y : x+y, neighborhood)
            
    # for each k-mer, check how many times it shows up in Text
    # with maximum mismatch, d
    
    # assign index and count to all possible k-mers  
    index = [0]*len(neighborhood_array)
    count = [0]*len(neighborhood_array) 
    for i in range(len(neighborhood_array)):
        pattern = neighborhood_array[i]
        index[i] = Pattern2Number(pattern)
        count[i] = 1
    sorted_index = sorted(index)
    
    # combine counts for same patterns 
    for i in range(1, len(neighborhood_array)):
        if sorted_index[i] == sorted_index[i-1]:
            count[i] = count[i-1]+1
    max_count = max(count)
    
    # using index of maximum counted pattern, withdraw pattern
    for i in range(len(neighborhood_array)):
        if count[i] == max_count:
            pattern = Number2Pattern(sorted_index[i], k)
            freq_pattern.append(pattern)
    return freq_pattern

# - Sample Input (Output: ATGT ACAT)
print('Sample Freq Words MisMatch & RC =', ' '.join(str(e) for e in FreqWordsMismatchRC('ACGTTGCATGTCGCATGATGCATGAGAGCT', 4, 1)))

# - Randomized Dataset
print('Randomized Freq Words MisMatch & RC =', ' '.join(str(e) for e in FreqWordsMismatchRC('TAGTCACTCCTCACACACTCAAATAGTCTAGTCTCTCCTCCTAGACACACTCCACTCACTAGTCACTCCAAAACTCCTCTAGTCTCCTCCAAAAAATAGACACAAAAAAAAAACAAATCCTCCTCTAGTCCAAATAGTCTCTCCTCCTAGAAATAGTAGAAAAAATAGTCTAGTCTCCAAAACACTAGAAAACACAAATCCACACTCCTAGACTAGTCAAATAGTAG', 5, 2)))
```

```python
# Skew Minimum Method

def MinimumSkew (genome):
  skew = 0
  skew_dict = {0:0}
  minimum_list = []
  for i in range (len(genome)):
    if genome[i]=='C':
      skew = skew - 1
      skew_dict[i] = skew
    elif genome[i]=='G':
      skew = skew + 1
      skew_dict[i] = skew
    else:
      skew_dict[i] = skew
  minimum_skew = min(skew_dict.values())
  for j in skew_dict:
    if skew_dict[j] == minimum_skew:
      minimum_list.append(j)
  return minimum_list;

MinimumSkew (genome = 'ATAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT')
```

```python
# Generate the d-Neighborhood of a String

# The d-neighborhood Neighbors(Pattern, d) is the set of all k-mers whose
# Hamming distance from Pattern does not exceed d.

# Input: A DNA string Pattern and an integer d.

# Return: The collection of strings Neighbors(Pattern, d).

def Neighbors(pattern, d):
    bases     = ['A', 'C', 'G', 'T']
    
    if d == 0:
        return (pattern)
    if len(pattern) == 1:
        return (bases)

    neighbors       = []
    suffixNeighbors = Neighbors(pattern[1:], d)
    for text in suffixNeighbors:
        if HammingDistance(pattern[1:], text) < d:
            for base in bases:
                neighbors.append(base + text)
        else :
            neighbors.append(pattern[0] + text)
    return(neighbors)

# Test
pattern = 'TGCAT'
d       = 2
d_neighbors = Neighbors(pattern, d)

print(d_neighbors)
print(len(d_neighbors))
```
